#  Data Cleaning

Checking for total null values

```python
data.isna().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
# Fill missing values with better strategies
data['director'].fillna('Not Available', inplace=True)
data['cast'].fillna('Not Available', inplace=True)
data['country'].fillna('Unknown', inplace=True)

# Fill rating with mode (most common rating)
data['rating'].fillna(data['rating'].mode()[0], inplace=True)

# Fill missing 'date_added' with the previous available date (forward fill)
data['date_added'].fillna(method='ffill', inplace=True)

# Optionally: handle release_year missing values if any
if data['release_year'].isnull().sum() > 0:
    data['release_year'].fillna(data['release_year'].median(), inplace=True)
```

## Univariate Analysis
#### Count of shows by type (movie vs tv show)

```python
data['type'].value_counts()
```

```python

# Count Movies vs TV Shows
df_pie = data['type'].value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(
    x=df_pie.values,
    labels=df_pie.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    colors=['#1f77b4','#2ca02c'],   # custom colors (blue & green)
    textprops={'fontsize':12, 'weight':'bold'},
    wedgeprops={'linewidth':2.0,'edgecolor':'white'}
)

plt.title('Movies vs TV Shows Ratio', fontsize=14, fontweight='bold')
plt.show()
```

```python
data['rating'].value_counts()
```

```python
fig = plt.figure(figsize=(16, 6))

# Create bins for countplot
bins = []
for i in range(1920, 2022, 10):
    if i == 2020:
        i = 2021
    bins.append(i)

df_count = data[['type', 'release_year']].copy()
df_count.loc[:, 'release_year_bins'] = pd.cut(df_count['release_year'], bins)

# Countplot with improved style
sns.set(style="whitegrid")
sns.countplot(
    data=df_count, 
    x='release_year_bins', 
    hue='type', 
    palette="Set2"
)

plt.xticks(rotation=45, fontsize=11)
plt.title('Distribution of Movies and TV Shows per 10 Years', fontsize=16, fontweight='bold')
plt.ylabel('Number of Shows Released', fontsize=12, fontweight='bold')
plt.xlabel('Release Year Range', fontsize=12, fontweight='bold')
fig.set_tight_layout(True)
plt.show()
```
